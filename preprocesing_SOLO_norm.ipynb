```python
import os

import open3d
import trimesh


def mesh_to_trimesh(mesh):
    """
    Converts a mesh object to a trimesh.
    """
    open3d.io.write_triangle_mesh("temp.ply", mesh)
    tri_mesh = trimesh.load_mesh("temp.ply")
    os.remove("temp.ply")
    return tri_mesh


def trimesh_to_mesh(tri_mesh):
    """
    Computes a trimesh object to a mesh object.
    """
    trimesh.exchange.export.export_mesh(tri_mesh, 'temp.ply', 'ply')
    mesh = open3d.io.read_triangle_mesh('temp.ply')
    os.remove("temp.ply")
    return mesh
```

```python
def translate_to_origin(mesh):
    """
    Translates the mesh,
    such that its centroid coincides with the coordinate-frame origin.
    """
    mesh.translate(-mesh.get_center())
    return mesh


def scale_to_unit(mesh):
    """
    Scales the mesh,
    such that it fits tightly in a unit-sized cube.

    The mesh must be located at the origin.
    """
    center = mesh.get_center()
    if center[0] > 0.001 or center[1] > 0.001 or center[2] > 0.001:
        raise ValueError(
            f'Mesh must be centered around the origin, not {center}'
        )
    factor = 1 / max(mesh.get_max_bound() - mesh.get_min_bound())
    mesh.scale(factor, center=True)
    return mesh
```

```python
def normalize_shape(mesh):
    """
    Normalizes the mesh by translating it to the origin and scaling it
    to fit within a unit-sized cube.
    """
    # Step 1: Translate the mesh to the origin (centroid at origin)
    mesh = translate_to_origin(mesh)

    # Step 2: Scale the mesh to fit within a unit cube
    mesh = scale_to_unit(mesh)

    return mesh
```
